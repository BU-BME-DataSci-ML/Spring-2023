```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
mu = 0.
sigma = 1.

dist = scipy.stats.norm(mu, sigma)
```

```python
xmin = -5.
xmax = 5.

x = np.arange(xmin, xmax, 0.01)
y = dist.pdf(x)

plt.plot(x, y, 'k-');
```

```python
xmin = -5.
xmax = 5.

x = np.arange(xmin, xmax, 0.01)
y = dist.cdf(x)

plt.plot(x, y, 'k-');

```
